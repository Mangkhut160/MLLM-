```python
def get_response(data):
    res = {'电池合规检测':'通过', '电池规格检测':'通过', '品牌检测':'通过', '日期检测':'通过', '电压检测':'通过', '图片规范检测':'通过'}
    data = str(data)
    for i in  data.split('|||'):
        if i == '使用电池不符合规范':
            res['电池合规检测'] = '驳回，使用电池不符合规范'
        elif i == '使用电池规格严重超标，不符合国标要求':
            res['电池规格检测'] = '驳回，使用电池规格严重超标，不符合国标要求'
        elif i == '电池品牌图片与填写品牌信息不一致':
            res['品牌检测'] = '驳回，电池品牌图片与填写品牌信息不一致'
        elif i == '电池生产日期与填写信息不一致':
            res['日期检测'] = '驳回，电池生产日期与填写信息不一致'
        elif i == '电池额定电压/额定容量与填写信息不一致':
            res['电压检测'] = '驳回，电池额定电压/额定容量与填写信息不一致'
        elif i == '非电池相关信息图片' or i == '非规范电池信息图片':
            res['图片规范检测'] = '驳回，上传图片信息非电池相关'
    return str(res) 
```

```python
df['response'] = df['label'].apply(lambda x: get_response(x))
df 
```

```python
def get_query(data):
    times,pv,brand,pi = data.split('$%$')
    prompt = f"""
    请针对输入的文本信息和两张图片信息进行相关内容的审核，具体的审核要求如下：
    1. 根据输入的品牌名称：{brand}与第一张图片中的品牌进行信息匹配
    2. 根据输入的电池的生产日期:{times}与第二张图片中的生成日期进行信息匹配
    3. 根据输入的电池的电压:{pv}、电流：{pi}与第二张图片中的电压、电流进行信息匹配
    4. 判断上传的两张图片是否与电池相关的图片、以及清晰度
    根据匹配的结果，如果符合要求则输出：通过，不符合则输出，驳回以及驳回的标签
    """.strip() 
    return prompt
```

```python
df['query'] = df['person_info'].apply(lambda x: get_query(x))
```

```python
import os
import shutil

def remove_brand_from_filenames(directory):
    """
    移除指定目录中PNG图片文件名中的"_brand"部分
    
    参数:
        directory (str): 包含图片文件的目录路径
    """
    # 遍历目录中的所有文件
    for filename in os.listdir(directory):
        # 只处理PNG文件
        if filename.lower().endswith('.png'):
            # 检查文件名中是否包含"_brand"
            if '_brand' in filename:
                # 构建新文件名（去掉"_brand"）
                new_filename = filename.replace('_brand', '')
                
                # 获取完整的旧文件路径和新文件路径
                old_path = os.path.join(directory, filename)
                new_path = os.path.join(directory, new_filename)
                
                # 重命名文件
                try:
                    os.rename(old_path, new_path)
                    print(f"已重命名: {filename} -> {new_filename}")
                except Exception as e:
                    print(f"重命名 {filename} 时出错: {e}")
            else:
                print(f"无需修改: {filename}")

if __name__ == "__main__":
    # 输入包含图片的目录路径
    target_directory = input("请输入包含图片的目录路径: ")
    
    # 检查目录是否存在
    if os.path.isdir('./brand_new/'):
        print(f"开始处理目录: {target_directory}")
        remove_brand_from_filenames(target_directory)
        print("处理完成!")
    else:
        print(f"错误: 目录 '{target_directory}' 不存在")
```
